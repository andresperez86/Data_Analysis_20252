- Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$
- Agrupe los datos por deporte calculando la media para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso en promedio?

- ¿Cuál es el pais con más participación femenina?

## Para responder esta pregunta tenemos varias forma de realizarlo

**Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.**

```python
athletes.query('gold >= 1 | silver >= 1 | bronze >= 1')
```

```python
athletes[athletes['bronze'] + athletes['gold'] + athletes['silver'] >= 1]
```

```python
athletes.query('gold + silver + bronze >= 1')
```

## **Para calcular la masa muscular por fila con una función como**

$$I_{m.m.} = \frac{Weight}{Height^2}$$

```python
# Podemos crear una funcion y llamarla para para cada uno de los atletas
def imm(row):
  ind_m_m = row.weight / (row.height ** 2)
  return ind_m_m

athletes.apply(imm, axis = 1)
```

## Agrupe los datos por deporte calculando la media para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso en promedio?

```python
athletes.groupby('sport').mean(numeric_only=True)[['weight']].sort_values(by = 'weight',ascending=False)
```

## **¿Cuál es el pais con más participación femenina?**
